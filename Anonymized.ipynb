```python
import pandas as pd

# Sample data
data = {
    'Age': [25, 34, 45, 52, 23, 30, 40, 50, 60, 70],
    'Zip-Code': ['123456', '234567', '345678', '456789', '567890', '678901', '789012', '890123', '901234', '012345'],
    'Diseases': ['Diabetes', 'Hypertension', 'Asthma', 'None', 'Cancer', 'None', 'Arthritis', 'None', 'Heart Disease', 'None']
}

# Create DataFrame
dff = pd.DataFrame(data)

# Output the DataFrame
print(dff)
```

```python
#Generalize Age to Age Range
def age_range(Age):
  if 20<=Age<=30:
    return '20-30'
  elif 31<=Age<=40:
    return '31-40'
  elif 41<=Age<=50:
    return '41-50'
  elif 51<=Age<=60:
    return '51-60'
  else:
    return '60+'
# Create the 'Age Range' column before dropping the 'Age' column
dff['Age Range']=dff['Age'].apply(age_range)
# Now you can safely drop the 'Age' column
dff = dff.drop(columns=['Age'])
print(dff)
```
